```python
X_column= crime_filtered['X']
Y_column = crime_filtered['Y']

```

```python
X_DIVS = 1000
Y_DIVS = 1000
```

```python
x_coord = (X_column + 122.55) // (0.2/ X_DIVS)
```

```python
y_coord = (Y_column - 37.65) // (0.2/ Y_DIVS)
```

```python
def get_coord(combined_val):
    return [((combined_val//X_DIVS)/Y_DIVS*0.2) + 37.65 , ((combined_val%X_DIVS)/X_DIVS*0.2) - 122.55 ]

def map_points(long, lat, X_DIVS, Y_DIVS):
    x_coord = (long + 122.55) // (0.2/ X_DIVS)
    y_coord = (lat - 37.65) // (0.2/ Y_DIVS)
    combined = x_coord + y_coord * X_DIVS
    return combined
    
    

#get_coord(627732.0)
print(get_coord(571651.0))
print(get_coord(532717.0))
print(get_coord(575651.0))
print(get_coord(670711.0))

```

```python
data = pd.Series(np.arange(X_DIVS * Y_DIVS))
for i_am in range(X_DIVS * Y_DIVS ):
    val = 0
    if(i_am in counts.keys()):
        val = counts[i_am]
    data[i_am] = val
data
        
    
```
